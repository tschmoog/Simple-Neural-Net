```python
import numpy as np 
import pandas as pd

def conv_string(a):
    float(a)
    

dataset = pd.read_csv('database.csv')
dataset.head(10)

```

```python
relevant = dataset[['Date', 'Time', 'Latitude','Longitude', 'Magnitude']].copy()
relevant.head(10)
```

```python
#Convert date string into DateTime objet

relevant['DateTime'] = pd.to_datetime(pd.Series(relevant['Date']))
relevant['DateTime2'] = pd.to_datetime(pd.Series(relevant['Time']))
relevant.head(10)
```

```python
#Use new datetime objects to create day and month features
relevant['Day'] = pd.DatetimeIndex(relevant['DateTime']).day
relevant['Month'] = pd.DatetimeIndex(relevant['DateTime']).month
relevant['Minute'] = pd.DatetimeIndex(relevant['DateTime2']).minute
relevant['Hour'] = pd.DatetimeIndex(relevant['DateTime2']).hour
relevant.head(10)
```

```python
for col in relevant:
    print (type(relevant[col][1]))
```

```python
final_data = relevant [['Latitude', 'Longitude', 'Month','Day','Hour', 'Minute', 'Magnitude']].copy()
final_data.head(10)
```

```python
for col in final_data:
    print (type(final_data[col][1]))
```

```python
#Seperate the data into input training matrix and output training matrix

input_matrix = final_data[['Latitude', 'Longitude', 'Month','Day','Hour', 'Minute']].copy()
output_matrix = final_data[['Magnitude']]
input_matrix.head(5)
```

```python
output_matrix.head(5)
```

```python
#Convert pandas dataframe into numpy array

training_input = input_matrix.as_matrix()
training_output = output_matrix.as_matrix()
```

```python
#Normalise input and output data
```

```python
import numpy as np

#Sigmoid function
def siggymoid(x, deriv = False):
    if(deriv == True):
        return x*(1-x)
    return 1/(1+np.exp(-x))

#Seed random numbers
np.random.seed(1)

#Initialize weights randomly with mean 0
weights = 2*np.random.random((6,1)) - 1

for iter in xrange(10000):
    #Forward propagation
    l0 = training_input
    l1 = siggymoid(np.dot(l0,weights))
    
    #Calculate error
    
    my_error = training_output - l1
    
    #Multiply how much we missed by the gradient of sigmoid 
    
    diff = my_error * siggymoid(l1, True)
    
    #Update weights
    weights += np.dot(l0.T, diff)
    
print "output"
print l1[-10:]
```
